```python
from pathlib import Path
import sys
import json

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT))

from src.data_loader import load_fridgewise_data
from src.models.cold_start_clustering import evaluate_cluster_model, compute_wcss_curve, ColdStartClusterRecommender
from src.models.case_based import build_fridge_cases, CaseBasedRecommender
from src.models.explainability import RecommendationExplainer
```

```python
data = load_fridgewise_data(ROOT)
metrics = evaluate_cluster_model(data.profiles, n_clusters=5)
print(f"silhouette={metrics.silhouette:.4f}, davies_bouldin={metrics.davies_bouldin:.4f}")
compute_wcss_curve(data.profiles).head()
```

```python
cases = build_fridge_cases(data)
cbr = CaseBasedRecommender(cases)
print(f"cases: {len(cases)}")
if not cases.empty:
    sample = cases.iloc[0]
    import json
    ings = json.loads(sample["fridge_ingredients"])
    print(cbr.explain_recommendation(ings, sample["dietary_preference"], sample["preferred_cuisine"]))
```

```python
explainer = RecommendationExplainer().fit(data)
print("SHAP surrogate fitted:", explainer._fitted)
```
